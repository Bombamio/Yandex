## Класс TeleBot

Класс `TeleBot` используется для любых взаимодействий с Bot API мессенджера Telegram. С помощью методов этого класса можно отправлять и принимать сообщения, обрабатывать команды, работать с файлами, управлять группами и каналами.

Начнём с самого простого: создадим экземпляр класса `TeleBot` и посредством специального метода этого класса отправим сообщение. 

При создании экземпляра `TeleBot` в него передаётся токен, аутентифицирующий бота в API Telegram. После создания объекта можно вызывать его методы — именно они отвечают за работу с API.

Создайте в директории проекта файл *kittybot.py* и добавьте в него код программы. Подставьте в код:

* `chat_id` своего аккаунта;

* токен, который вы получили от @Botfather.

```python
# kittybot/kittybot.py

from telebot import TeleBot

# Укажите токен, 
# который вы получили от @Botfather при создании бот-аккаунта:
bot = TeleBot(token='<token>')
# Укажите id своего аккаунта в Telegram:
chat_id = <chat_id>
message = 'Вам телеграмма!'
# Вызываем метод send_message, с помощью этого метода отправляются сообщения:
bot.send_message(chat_id, message)
```
