## Рекурсивная функция

Рекурсивная функция `disassemble_matryoshka()` примет на вход объект матрёшки и проверит, есть ли в переданном объекте вложенная матрёшка.

Если вложенная матрёшка есть — напечатает сообщение `Разобрали матрёшку размера <размер>, разбираем дальше`! и рекурсивно вызовет саму себя, передав аргументом вложенный объект матрёшки.
Если вложенной матрёшки нет — напечатает сообщение `Все матрёшки разобраны! Размер последней матрёшки: <размер>` и завершит работу.

```python
class Matryoshka:

    def __init__(self, size, item=None):
        self.size = size
        self.inner_item = item


def disassemble_matryoshka(matryoshka):
    """Функция разборки матрёшки."""
    # Получаем вложенную матрёшку.
    inner_item = matryoshka.inner_item
    # Если вложенной матрёшки нет, значит, это была самая маленькая матрёшка 
    # и работа закончена.
    if inner_item is None:
        print(f'Все матрёшки разобраны! Размер последней матрёшки: {matryoshka.size}')
        return
    # Если вложенная матрёшка есть, печатаем информационное сообщение...
    print(f'Разобрали матрёшку размера {matryoshka.size}, разбираем дальше!')
    # ...и рекурсивно вызываем ту же функцию,
    # но аргументом в неё передаём объект, вложенный в текущий:
    disassemble_matryoshka(inner_item)

if __name__ == '__main__':
    # Создаём экземпляр матрёшки из трёх фигурок.
    big_matryoshka = Matryoshka('L', Matryoshka('M', Matryoshka('S')))
    # Передаём эту матрёшку на разборку.
    disassemble_matryoshka(big_matryoshka)
```

В последней строчке функции `disassemble_matryoshka()` вызывается та же функция, которая её и вызывает. Это и есть **рекурсивный вызов функции**: функция многократно вызывает сама себя.

Когда последний из «вложенных» вызовов выполнится, последняя вызванная функция вернёт результат своей работы в предпоследнюю — и завершит работу. 

Предпоследняя вызванная функция передаст свой результат ещё на уровень выше и тоже «выключится». 

Внешняя, самая первая функция будет работать до тех пор, пока не получит ответ от той функции, которую она сама вызвала, — от второй.
