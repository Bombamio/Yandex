Понятие «структура данных» не привязано к определённому языку программирования — ведь алгоритмы можно реализовать на любом языке. В разных языках одна и та же структура данных может быть по-разному реализована, и эти реализации могут называться по-разному. 

Структуры данных отличаются друг от друга способами хранения данных и способами доступа к ним. А в конкретных языках программирования эти структуры могут быть реализованы, например, в виде какого-то **типа** данных.
