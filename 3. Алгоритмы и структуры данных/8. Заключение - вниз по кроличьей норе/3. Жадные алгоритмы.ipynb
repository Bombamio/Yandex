«Жадным» называют такой алгоритм, который выбирает оптимальный ход лишь для текущего этапа, не учитывая всей картины. Алгоритм «жадничает», выбирая лучшее из возможного «здесь и сейчас».

Жадные алгоритмы исходят из того, что если на каждом шаге решение будет оптимальным, то и общий результат окажется наилучшим. Однако это не обязательно так.

Жадный алгоритм может ошибаться, но может возвращать и оптимальный ответ: например, при определённых сочетаниях банкнот жадный алгоритм будет выдавать оптимальное решение. Если в наличии будут банкноты по 1, 5, 10 и 25 марсиков, то жадный алгоритм справится с задачей наилучшим образом, независимо от требуемой суммы. В других условиях решение может оказаться неоптимальным.

## Ленивый коммивояжёр

Если применить жадный алгоритм к задаче коммивояжёра, то на каждом шаге путешественник станет выбирать самый короткий путь из доступных оставшихся. 

Получится не самый плохой маршрут, а при удачном стечении обстоятельств жадный алгоритм может построить и оптимальный маршрут, но в общем случае результат будет далёк от наилучшего. Лень и жадность — плохие помощники для коммивояжёра.

Вычислить оптимальный маршрут для ста городов невозможно технически. Лучше получить неоптимальный результат сейчас, чем оптимальный — никогда. И жадные алгоритмы — один из способов получить такой результат.

В ряде задач жадные алгоритмы дают оптимальные или близкие к оптимальным решения. Есть и бонус: как правило, они довольно просты в реализации, и это ценно для разработчиков.
