# Операторы сравнения

- больше >,
- больше или равно >=,
- меньше <,
- меньше или равно <=,
- равно ==,
- не равно !=.

- `==` Равно: проверяет, что два значения равны.
> Например: `x == 5` вернёт `True`, если x равно 5.

- `!=` Не равно: проверяет, что два значения не равны.
> Например: `x != 5` вернёт `True`, если x равно 6.

- `<` Строго меньше: проверяет, что левое значение меньше правого.
> Например: `x < 5` вернёт `True`, если x меньше 5.

- `>` Строго больше: проверяет, что левое значение больше правого.
> Например: `x > 5` вернёт `True`, если x больше 5.

- `<=` Меньше или равно: проверяет, что левое значение меньше правого или равно ему.
> Например: `x <= 5` вернёт `True`, если x меньше или равно 5.

- `>=` Больше или равно: проверяет, что левое значение больше правого или равно ему.
> Например: `x >= 5` вернёт `True`, если x больше или равно 5.

# Булевы значения
Значения `True` и `False` принадлежат типу bool. Других значений, кроме `True` и `False`, у данных этого типа нет. Этот тип называется «логическим».

## Цепочки сравнений

> Вместо того чтобы писать два отдельных выражения, можно использовать `цепочку сравнений`.

```python
current_temperature = 19

print(10 < current_temperature < 20)
```

## Приведение булевых значений к числам

Обычно в программировании ложные выражения обозначают *нулём*, а истинные — *единицей*. Целочисленные значения можно привести к булевым значениям, а булевы — к целочисленным.

Любые числовые значения, кроме 0, приводятся к `True`.

```python
bool_true = True     # Булево значение.
bool_false = False   # Булево значение.
int_null = 0         # Число.
int_one = 1          # Число.
int_100 = 100        # Число.
int_negative = -100  # Число.
float_pi = 3.14      # Число с плавающей точкой.

print('Вместо True -', int(bool_true))   # Приводим bool к числу.
print('Вместо False -', int(bool_false)) # Приводим bool к числу.

print('Вместо 0 -', bool(int_null))         # Приводим число к bool.
print('Вместо 1 -', bool(int_one))          # Приводим число к bool.
print('Вместо 100 -', bool(int_100))        # Приводим число к bool.
print('Вместо -100 -', bool(int_negative))  # Приводим число к bool.
print('Вместо 3.14 -', bool(float_pi))      # Приводим число к bool.
```

```python
jar_of_vegetables = None

# Так лучше не делать:
print(jar_of_vegetables == None)  # Оператор == не годится для проверки на None.

# Проверять, равна ли переменная None,
# надёжнее через оператор is:
print(jar_of_vegetables is None)
# True
```

# Что в итоге

- Тип данных **bool** — логический. Данные типа **bool** могут принимать лишь два значения: `True` (истина) и `False` (ложь). Никаких «может быть» или «не знаю».
- **Операции сравнения** — это проверка выражения на истинность. При операциях сравнения возвращается значение булева типа.
- **Комбинированные операторы сравнения** — «больше или равно» и «меньше или равно» — возвращают `True`, если хотя бы одно из заявленных «предположений» истинно.
- *Значения булева типа можно привести к числовому типу*: `True` будет преобразовано в `1`, `False` — в `0`. Числа тоже можно привести к типу `bool`: `0` преобразуется в `False`, остальные числа — в `True`.
