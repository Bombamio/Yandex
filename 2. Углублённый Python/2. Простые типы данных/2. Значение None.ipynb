Тип `NoneType` — это отдельный тип данных, к этому типу относится лишь одно значение: `None`. 

Например, данные типа «число» могут быть бесконечно разнообразны (`5`, `-101`, `0`, `42` — это всё значения для типа `int`), а в `NoneType` такого разнообразия нет и быть не может: «ничего» — оно всегда одно и то же, `None`. Это отсутствие какого бы то ни было значения.

`None` означает «ничего», и это «ничего» можно присвоить переменной.

```python
# Переменная jar_of_vegetables содержит "ничего".
jar_of_vegetables = None
```

# Зачем этот None

Самый простой пример: разработчику надо объявить переменную, которая, по планам, точно понадобится в коде, но пока что непонятно, какое значение ей присвоить. Это примерно та же ситуация, что и с объявлением пустых функций: как нельзя объявить функцию без тела — точно так же нельзя объявить переменную без значения. В этой ситуации вместо значения можно указать `None`.

Можно применять `None` как значение по умолчанию для параметров функций. Это удобно в тех случаях, когда необходимо разрешить вызовы функции без явно указанного аргумента.

# Что в итоге

- Переменные с типом данных `NoneType` могут содержать лишь `None`, никаких других вариантов нет. `None` символизирует отсутствие значения как такового.
- Любая функция, в которой не применена инструкция `return`, возвращает `None`.
- `None` можно использовать для дефолтного значения параметров функции или для объявления переменной.
