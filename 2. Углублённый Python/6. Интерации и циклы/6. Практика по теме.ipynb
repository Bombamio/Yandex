```python
def print_pack_report(starting_value):
    # Объявите диапазон от starting_value до 1 включительно
    # и переберите его в цикле:
    if starting_value < 1:
        return 'Введите число больше 1'
        
    for number in range(starting_value, 0, -1):
        # Проверьте, делится ли текущий элемент
        # на 3, на 5 и на 3 и 5 одновременно.
        # В зависимости от результата проверки
        # напечатайте нужную фразу
        if number % 3 == 0 and number % 5 == 0:
            print(f'{number} - расфасуем по 3 или по 5')
        elif number % 3 == 0 or number % 5 == 0:
            if number % 3 == 0:
                print(f'{number} - расфасуем по 3')
            else:
                print(f'{number} - расфасуем по 5')
        else:
            print(f'{number} - не заказываем!')


print_pack_report(31)
```

```python
def count_canisters(temperatures_per_day):
    hot_days_counter = 0
    for temperature in temperatures_per_day:
        if temperature >= 30:
            hot_days_counter += 1
    # Допишите функцию.
    return hot_days_counter


forecast_temperatures = [26, 28, 30, 31, 29, 31, 28, 26]
# Вызовите функцию и напечатайте результат в нужном формате.
print(f'Нужно канистр: {count_canisters(forecast_temperatures)}')
```

```python
def print_multiplication_table():
    # Напишите код, который напечатает таблицу умножения.
    for x in range(1, 10):
        for y in range(1, 10):
            print(f'{x} * {y} = {x * y}')
        print('-' * 10)


print_multiplication_table()
```

```python
def analyze_results(data):
    # Получите названия всех команд и отсортируйте их по алфавиту.
    races_keys = list(data[0].keys())
    races_keys.sort()

    print('Команды, участвовавшие в чемпионате:')
    for name in races_keys:
        print(f'* {name}')
    
    races_keys = dict.fromkeys(races_keys, 0)
    
    for num in range(len(data)):
        for key in data[num].keys():
            races_keys[key] += data[num][key]
    
    for team, score in races_keys.items():
        if score == max(races_keys.values()):
            print(f'В чемпионате победила команда {team} с результатом {score} баллов')
            break
    

races_data = [
    {'Ferrari': 20, 'Mercedes': 5, 'Aston Martin': 10, 'Williams': 15},
    {'Mercedes': 15, 'Aston Martin': 20, 'Ferrari': 10, 'Williams': 0},
    {'Ferrari': 20, 'Williams': 15, 'Aston Martin': 10, 'Mercedes': 5}
]

analyze_results(races_data)
```
