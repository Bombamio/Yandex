## 1. Начинающий Python-разработчик решил выяснить, какие навыки нужны для того, чтобы из начинающего разработчика превратиться в успешного.

В сети нашёлся перечень дополнительных навыков, которыми владеют другие питонисты.

Данные приведены в процентах. Кто-то из разработчиков владеет сразу несколькими дополнительными навыками, поэтому сумма значений — заметно больше ста.

Напишите функцию `analyze_skills()`, которая 
- найдёт минимальное значение среди перечисленных переменных,
- найдёт максимальное значение среди перечисленных переменных,
- напечатает две строки:

```python
Доля питонистов, у которых есть наименее популярный навык (в %): <мин.число>
Доля питонистов, у которых есть наиболее популярный навык (в %): <макс.число>
```

- Название навыка печатать не нужно. 
- Функция не должна возвращать никакого значения.
- Многоточия из прекода нужно удалить.

```python
bash = 31
c_and_c_plus_plus = 29
c_sharp = 11
html_css = 36
java = 19
javascript = 37
sql = 34

def analyze_skills():
    # Среди всех переменных найдите минимальное и максимальное значение.
    min_number = min(bash, c_and_c_plus_plus, c_sharp, html_css, java, javascript, sql)
    max_number = max(bash, c_and_c_plus_plus, c_sharp, html_css, java, javascript, sql)
    
    # Напечатайте фразы, описанные в задании (две строки).
    print(f'Доля питонистов, у которых есть наименее популярный навык (в %): {min_number}')
    print(f'Доля питонистов, у которых есть наиболее популярный навык (в %): {max_number}')

# Не удаляйте вызов функции.
analyze_skills()
```

## 2. Начинающий, но в будущем успешный разработчик никак не угомонится. «А правильно ли было выбрать Python? Может, надо было заняться изучением Java или даже C#? Посмотрю статистику по вакансиям».

В Интернете нашлось исследование: всемирный рейтинг языков программирования по числу вакансий. Вакансий много, их количество указано в тысячах:

```python
c_sharp = 375
java = 546
javascript = 915
... 
```

Допишите функцию `analyze_jobs()`, в ней обработайте переменные, в которых хранится количество вакансий:
- подсчитайте суммарное количество вакансий для перечисленных языков программирования,
- подсчитайте процент вакансий для Python от числа всех исследованных вакансий,
- с помощью встроенной функции `round()` округлите вычисленный процент до двух знаков после запятой (до сотых долей),
- дважды вызовите `print()`, чтобы напечатать фразы:

```python
Общее число исследованных вакансий, в тысячах: <сумма_всех_вакансий>
Вакансии для Python-разработчиков, в %: <процент_вакансий_для_Python>
```

```python
# Количество вакансий для различных языков программирования:
c_sharp = 375
java = 546
java_script = 915
php = 288
python = 603

def analyze_jobs():
    # Вычислите общее количество исследованных вакансий.
    total_jobs = c_sharp + java + java_script + php + python
    
    # Вычислите процент вакансий для Python от общего числа вакансий
    # и округлите результат до двух знаков (до сотых долей):
    python_percent = python / total_jobs * 100
    
    # Напечатайте фразы, описанные в задании (две строки).
    print(f'Общее число исследованных вакансий, в тысячах: {total_jobs}')
    print(f'Вакансии для Python-разработчиков, в %: {round(python_percent, 2)}')

analyze_jobs()
```

Напишите функцию `get_dumplings_recommendation()`, которая будет случайным образом возвращать целые числа от `10` до `20` включительно. 

Для работы вам понадобится функция, возвращающая случайное целое число. Импортируйте её из модуля random и задайте ей псевдоним `rnd`.

Вызовите функцию и напечатайте сообщение:

```python
Оптимальным количеством пельменей на сегодня будет <число>
```
